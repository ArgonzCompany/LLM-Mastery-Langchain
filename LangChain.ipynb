```python
import torch
import os
import getpass

from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline, BitsAndBytesConfig
from langchain_huggingface import HuggingFacePipeline

from langchain_core.messages import SystemMessage
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

**Loading LLM via pipeline**

```python
model_id = "microsoft/Phi-3-mini-4k-instruct"
```

```python

quantization_config = BitsAndBytesConfig(
    load_in_4bit = True,
    bnb_4bit_quant_type = "nf4",
    bnb_4bit_use_double_quant = True,
    bnb_4bit_compute_dtype = torch.bfloat16
)
```

```python
model = AutoModelForCausalLM.from_pretrained(model_id, quantization_config = quantization_config)
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
pipe = pipeline(
    model = model,
    tokenizer = tokenizer,
    task = "text-generation",
    temperature = 0.1,
    max_new_tokens = 500,
    do_sample = True,
    repetition_penalty = 1.1,
    return_full_text = False
)
```

```python
llm = HuggingFacePipeline(pipeline = pipe)
```

```python
input = "What is AI?"
output = llm.invoke(input)
print(output)
```

**About LCEL**

```python
from langchain_core.output_parsers import StrOutputParser

chain_str = chain | StrOutputParser()
```

```python
chain_str.invoke({"topic": "quantum computing", "size": "1 phrase"})
```

**Custom Functions with Runnables**

```python
from langchain_core.runnables import RunnableLambda
count = RunnableLambda(lambda x: f"Words: {len(x.split())}\n{x}")
```

```python
chain = prompt | llm | StrOutputParser() | count
chain.invoke({"topic": "AI", "size": "1 phrase"})
```

**Streaming**

```python
for chunk in chain_str.stream({"topic": "black holes", "size": "1 paragraph"}): #The output streaming words by words step by step
  print(chunk, end="")
```

**Access models via Hugging Face Hub**

```python
from langchain_huggingface import HuggingFaceEndpoint

os.environ["HUGGINGFACEHUB_API_TOKEN"] = getpass.getpass()
```

```python
model_id = "alpindale/Llama-3.2-3B-Instruct"
```

```python
llm_hub = HuggingFaceEndpoint(
    repo_id = model_id,
    temperature = 0.1,
    max_new_tokens = 512,
    model_kwargs = {
        "max_length": 64
    }
)

response = llm_hub.invoke("What are the names of the planets in the solar system?")
print(response)
```
